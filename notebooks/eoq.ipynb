```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv("D:/Projects/inventory_optimization/data/SalesFINAL12312016.csv")

prices = pd.read_csv("D:/Projects/inventory_optimization/data/2017PurchasePricesDec.csv")
```

```python
sales["SalesDate"] = pd.to_datetime(sales["SalesDate"])
```

```python
sales.columns, prices.columns
```

```python
product_demand = (
    sales.groupby("Description")["SalesDollars"]
    .sum()
    .sort_values(ascending=False)
)

top_product = product_demand.index[0]
top_product
```

```python
annual_demand = product_demand.loc[top_product]
annual_demand
```

The product shows a high annual demand of over 43,000 units, indicating that it is a fast-moving item requiring frequent replenishment.

```python
unit_cost = (
    prices[prices["Description"] == top_product]["PurchasePrice"]
    .mean()
)

unit_cost
```

With a relatively low unit cost (₹8.18), holding excessive inventory may appear inexpensive per unit, but at scale it can significantly increase total carrying costs.

```python
ordering_cost = 500          # ₹ per order (assumed)
holding_cost_rate = 0.20     # 20% of unit cost per year
```

```python
holding_cost = holding_cost_rate * unit_cost
holding_cost
```

```python
EOQ = np.sqrt((2 * annual_demand * ordering_cost) / holding_cost)
EOQ
```

The EOQ of approximately 5,150 units suggests that ordering in moderate batch sizes minimizes the combined cost of ordering and inventory holding. Ordering significantly more or less than this quantity would increase total inventory costs.

At the EOQ level, the business would place approximately 8–9 orders per year (43,423 / 5,152), enabling predictable and efficient replenishment cycles.
